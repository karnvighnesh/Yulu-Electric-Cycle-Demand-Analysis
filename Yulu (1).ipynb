```python
Q1 = data['count'].quantile(0.25)
Q3 = data['count'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Count of outliers
outliers = data[(data['count'] < lower_bound) | (data['count'] > upper_bound)]
print("Number of outliers:", len(outliers))
```

```python
data_cleaned = data[(data['count'] >= lower_bound) & (data['count'] <= upper_bound)]
print("Original shape:", data.shape)
print("Cleaned shape:", data_cleaned.shape)
```

```python
from scipy.stats import f_oneway

season1 = data_cleaned[data_cleaned['season'] == 1]['boxcox_count']
season2 = data_cleaned[data_cleaned['season'] == 2]['boxcox_count']
season3 = data_cleaned[data_cleaned['season'] == 3]['boxcox_count']
season4 = data_cleaned[data_cleaned['season'] == 4]['boxcox_count']

f_stat, p_val = f_oneway(season1, season2, season3, season4)

print(f"F-Stat: {f_stat}, P-value: {p_val}")
```

Since p-value  0.05, we strongly reject H.

This result is extremely significant.

There is a statistically significant difference in electric cycle demand across seasons.

This means seasonality plays a strong role in how much users engage with Yulus services.

Krushkal Test

```python
from scipy.stats import kruskal
season1 = data_cleaned[data_cleaned['season'] == 1]['count']
season2 = data_cleaned[data_cleaned['season'] == 2]['count']
season3 = data_cleaned[data_cleaned['season'] == 3]['count']
season4 = data_cleaned[data_cleaned['season'] == 4]['count']

h_stat, p_val = kruskal(season1, season2, season3, season4)

print(f"Kruskal-Wallis H-statistic: {h_stat:.4f}, P-value: {p_val:.4f}")
```

Since p-value < 0.05, we reject the null hypothesis.

 There is a statistically significant difference in median electric cycle demand across seasons, even when using a non-parametric test.

H0 : Mean demand is the same across all weather conditions.

Ha : At least one weather type has a different mean demand.

```python
weather1 = data_cleaned[data_cleaned['weather'] == 1]['boxcox_count']
weather2 = data_cleaned[data_cleaned['weather'] == 2]['boxcox_count']
weather3 = data_cleaned[data_cleaned['weather'] == 3]['boxcox_count']
weather4 = data_cleaned[data_cleaned['weather'] == 4]['boxcox_count']

f_stat, p_val = f_oneway(weather1, weather2, weather3, weather4)
print(f"F-Stat: {f_stat}, P-value: {p_val}")
```

Since p-value  0.05, you strongly reject the null hypothesis.

 There is a statistically significant difference in electric cycle demand across different weather conditions.

H0: Season and Weather are independent (i.e., the weather distribution does not vary by season).

Ha : Season and Weather are dependent (i.e., certain weather types are more likely in specific seasons).

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data_cleaned['season'], data_cleaned['weather'])
chi2, p_val, dof, expected = chi2_contingency(contingency_table)
print(f"Chi2: {chi2}, P-value: {p_val}")
```

Since p-value  0.05, we reject the null hypothesis.

 There is a significant association between season and weather  certain weather conditions occur more frequently in specific seasons.

Key Insights:

1.Demand Distribution is Non-Normal:

The original rental count variable is right-skewed and leptokurtic.

Even after Box-Cox transformation (optimal  = 0.34), the Shapiro-Wilk test indicated a p-value  0.0 due to the large sample size, although the Q-Q plot showed improved normality.


2.Working Day Impact:

Two-sample t-test (Box-Cox transformed data) gave p-value = 0.0199, indicating a significant difference in electric cycle demand between working and non-working days.

Interpretation: Working days have slightly different demand, possibly due to commuting patterns.


3.Seasonal Effect:

ANOVA on Box-Cox transformed data gave a highly significant p-value < 1e-125.

Kruskal-Wallis (non-parametric) also confirmed this (p-value = 0.0000).

Insight: Season has a strong influence on demand, likely due to weather patterns, holidays, and temperature shifts.

4.Weather Impact:

ANOVA across weather conditions showed a significant difference (p-value  1.84e-37).

Interpretation: Bad weather significantly reduces demand.

5.Season-Weather Association:

Chi-square test on season vs weather yielded p-value = 3.64e-07, meaning the two are not independent.

Insight: Weather conditions are seasonally driven, which may lead to collinearity in models using both features
